# check for outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = df.select_dtypes(include=['int64','float64']).columns
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[num_cols], orient='h')
plt.title('outliers across all numerical features')
plt.show()
```

```python
for col in num_cols:
    sns.boxplot(data=df[col], orient='h')
    plt.title(col)
    plt.show()
```

```python

def iqr_capping(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # Define bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Cap outliers
        df[col] = df[col].clip(lower_bound, upper_bound)
    return df


df_cleaned = iqr_capping(df.copy(), num_cols)
```

```python
for col in num_cols:
    sns.boxplot(data=df_cleaned[col], orient='h')
    plt.title(col)
    plt.show()
```

```python
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".1f", cmap="YlGnBu")
```

```python
def plot_continuous_target_vs_features(df, target_col='review_score', cols_per_row=3):
    numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
    numerical_features = numerical_features[numerical_features != target_col]
    
    n_features = len(numerical_features)
    n_rows = (n_features // cols_per_row) + (1 if n_features % cols_per_row else 0)
    
    plt.figure(figsize=(15, 5 * n_rows))
    
    for i, feature in enumerate(numerical_features, 1):
        plt.subplot(n_rows, cols_per_row, i)
        sns.scatterplot(x=feature, y=target_col, data=df, alpha=0.6)
        plt.title(f'{target_col} vs {feature}')
        
    plt.tight_layout()
    plt.show()
```
